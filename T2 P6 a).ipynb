```python
import numpy as np
import matplotlib.pyplot as plt
import camb
from camb import model
```

```python
H0 = 67.4 # km/s/Mpc
omega_m = 0.315
omega_ch2 = 0.120
omega_bh2 = 0.0224 
ns = 0.965
tau = 0.054
```

```python
pars = camb.CAMBparams();
pars.set_cosmology(H0=H0, ombh2=omega_bh2, omch2=omega_ch2, mnu=0.06, omk=0);
pars.set_matter_power(redshifts=[0.0,0.8, 5,10,50], kmax=4.0);
```

```python
results = camb.get_results(pars)
```

```python
kh, z, pk = results.get_matter_power_spectrum(minkh=1e-4, maxkh=2, npoints=200)
```

```python
for i, (redshift) in enumerate(zip(z)):
    plt.loglog(kh, pk[i, :])
plt.xlabel('k/h Mpc')
```
